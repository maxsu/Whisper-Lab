```python
from pathlib import Path
from whisperlab.run_whisper import run_whisper, WhisperRequest
from whisperlab import logger

# Data Sources
FILE_0 = Path("../audio/the_most_dangerous_game.mp3")
assert(FILE_0.exists())

# Configure the logger
LOG_0 = Path("../audio/whisper.TheMostDangerousGame.log")
logger.config(verbose=True, log_file=LOG_0)

logger.info("Starting the Whisper Lab Demo")

# Configure the Whisper Request
request = WhisperRequest(
    audio_file=FILE_0,
)

# Run the Transcription
result = run_whisper(request)
```
